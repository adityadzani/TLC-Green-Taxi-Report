#### Combining all the record

```python
import os
import glob 
import pandas as pd
os.chdir('/Data Science/EDA/TLC Green Taxi Trip Record Data')
```

```python
extension= 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
data= pd.read_csv('green_taxi_2020.csv')
```

```python
sample_data = data.sample(frac= 0.3, random_state=1)
```

```python
sample_data.shape
```

```python
sample_data.head()
```

```python
sample_data.drop(columns= 'DOLocationID', inplace= True)
```

```python
sample_data['VendorID'] = sample_data['VendorID'].replace({1:'CCT', 2:'VeriFone Inc.'})
sample_data['RatecodeID'] = sample_data['RatecodeID'].replace({1:'Standard', 2:'JFK', 3:'Newark', 4:'Nassau', 5:'Negotiated', 6:'Group Ride'})
sample_data['payment_type'] = sample_data['payment_type'].replace({1:'Credit Card', 2: 'Cash', 3:'No Charge', 4:'Dispute', 5:'Unknown', 6:'Voided trip'})
sample_data['trip_type'] = sample_data['trip_type'].replace({1:'Street-hail', 2:'Dispatch'})
```

```python
sample_data.head()
```

```python
sample_data.shape[0]
```

```python
location = pd.read_csv('taxi+_zone_lookup.csv', sep= ';')
```

```python
location.head()
```

```python
location.drop(columns= ['Zone', 'service_zone'], inplace= True)
```

```python
sample_data = sample_data.merge(location, left_on= 'PULocationID', right_on='LocationID')
```

```python
sample_data['Borough'].value_counts()
```

```python
sample_data.head()
```

```python
sample_data.drop(columns=['PULocationID', 'LocationID'], inplace= True)
```

```python
sample_data.head()
```

```python
sample_data = sample_data.sample(frac=0.95, random_state=1)
```
